```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv sin indice
df = pd.read_csv('cuentas_credicel.csv', encoding= 'latin') 
```

a) Todas las columnas excepto: ”folio”, “folio_solicitud”, “Unnamed: 33”, “Unnamed: 34”, “Unnamed: 35” y “Unnamed: 36

```python
df1= df.iloc[ :, [1] +  list(range(3, 33))]  #Filas combinadas
df1
```

b) Todas las filas multimplos de 4 

```python
df2 = df1.iloc[0::4]
df2
```

c) Todos los registros cuya marca sea igual a: ”ALCATEL”, “MOTOROLA”, “NOKIA”, “ZTE” y “SAMSUNG”

```python
#Filtro Y
df3=df2[(df2["marca"] == "ALCATEL") | (df2["marca"] == "MOTOROLA") |  (df2["marca"] == "ZTE") | (df2["marca"] == "NOKIA") |  (df2["marca"] == "SAMSUNG")]
df3
```

d) Todos los registros cuyo enganche se encuentre entre: ”$500 y $2000”

```python
#Filtro Y
df4=df3[(df3["enganche"] >= 500) & (df3["enganche"] <= 2000)]
df4
```

  e) Todos los registros cuya categoría en riesgo: NO sea “Cancelado” ni “Fraude”

```python
#Filtro not
df5 =df4[~(df4["riesgo"] == "Cancelado") & ~(df4["riesgo"] == "Fraude")]
df5
```

 f) Todos los registros cuya fecha : Sea posterior a “15/11/2021”

```python
#Filtro por objeto
df6= df5[df5["fecha"] > "2021-11-15"]
df6
```

```python
#Convertir archivo filtrado a CSV
df6.to_csv("Act7_Ej3.csv")
```
